```python
# This is a tool for visualizing trees
import svgling
from svgling.figure import Caption, SideBySide

# This is a dependency parser and its visualizer
import spacy
from spacy import displacy
nlp = spacy.load('en_core_web_trf')

# These are some helper functions to keep the interactive code simple
def vis(sentence: str) -> None:
    doc = nlp(sentence)
    displacy.render(doc, style="dep")

def vis_trees(brackets_1, brackets_2, caption: str, font_size=20) -> None:
    tree_1 = svgling.draw_tree(brackets_1, font_size=font_size)
    tree_2 = svgling.draw_tree(brackets_2, font_size=font_size)
    display(Caption(SideBySide(*[tree_1, tree_2]), caption, font_size=font_size+2))
```

# The students saw the man with the telescope.

Many sentences are ambiguous. Sometimes, this is because there is more than one way to draw that sentence's tree.



Draw a tree for each possible interpretation of the sentence.

> The students saw the man with the telescope.

```python
tree_1 = ("S",
              ("NP",
                    ("Det", "the"),
                    ("N", "students")),
              ("VP",
                       ("V", "saw"),
                       ("NP", 
                            ("NP",
                                 ("Det", "the"),
                                 ("N", "man")),
                            ("PP",
                                 ("P", "with"),
                                 ("NP",
                                      ("Det", "the"),
                                      ("N", "telescope"))))))

tree_2 = ("S",
              ("NP",
                   ("Det", "the"),
                   ("N", "students")),
              ("VP",
                   ("V", "saw"),
                   ("NP",
                        ("Det", "the"),
                        ("N", "man")),
                   ("PP",
                        ("P", "with"),
                        ("NP",
                             ("Det", "the"),
                             ("N", "telescope")))))

vis_trees(tree_1, tree_2, "What are the distinct meanings represented by these trees?")
```
